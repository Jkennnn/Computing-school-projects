```python
def NRIC(n):
    # Valid prefixes
    prefixes = ['S', 'T', 'M', 'F', 'G']

    while True:
        # Length check
        if (len(n))!=9:
            n = input("NRIC must be 9 characters long. Please input an NRIC: ")
            continue

        # Prefix check
        if n[0] not in prefixes:
            n = input("Prefix must be S, T, M, F, or G. Please input an NRIC: ")
            continue

        # Digits check (positions 1-7)
        all_digits = True
        for char in n[1:8]:
          if not char.isdigit(): #char.isdigit==(false)
            all_digits=False
          break


        if not all_digits: #all_digits==False
            n = input("Positions 2 to 8 must be digits. Please input an NRIC: ")
            continue

        # Suffix check
        if not n[8].isalpha():
            n = input("Last character must be an alphabet. Please input an NRIC: ")
            continue

        # If all checks pass
        break

    return "NRIC is valid."

n = input("Please input an NRIC: ")
print(NRIC(n))
```
