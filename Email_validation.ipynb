```python

# List to store registered emails
registered_emails = ["abc@mail.com", "def@mail.com", "ghi@mail.com"]

print("=== Application Form ===")

# Validate Name
while True:
    space=False
    name = input("Enter your full name (First Name Last Name): ")
    for i in range (len(name)):
      if name[i]==" ":
        space=True
        break
    if space==True:
      break
    else:
      print("name must include full name with spacing.")


# Validate Age
while True:
  ages=False
  age = input("Enter your age (must be 16 or older): ")
  if age.isdigit() and int(age)>=16:
    ages=True
    break
  else:
    print("age must be a number and must be 16 or above.")




# Validate Email
while True:
  email = input("Enter your email: ")
  at=False
  dot=False
  if email in registered_emails:
    print("email cannot be a registered email.")
  else:
    for i in range(len(email)):
      if email[i]=="@":
        at=True
      if email[i]==".":
        dot=True
    if at==True and dot==True:
      break
    else:
      print("email must require and @ and a dot.")








# Display successful submission
print("\n=== Application Submitted Successfully ===")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Email: {email}")
```
