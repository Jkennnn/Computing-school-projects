```python
#Task 5.5
def total_cost(cost):
  total = cost * 1.09 # adds 9 percent to the original cost
  return total # return total cost after gst

def discount(cost):
  if 50<=cost<100:
    new_cost=cost * 0.95
  elif cost >= 100:
    new_cost= cost * 0.90
  else:
    new_cost=cost # if total cost is less than 50
  return new_cost # returns the discounted price

def rewards_point(cost):
  points=cost*3
  return points #returns the reward points

def vouchers(cost,first_name):
  if 25<cost<=50:
    code=(first_name[0:3]+"05PERCENT")
  elif cost>50:
    code=(first_name[0:3]+"10PERCENT")
  return code #returns voucher code

name=input("Enter your name: ")

cost=float(input("Enter your cost: "))

print("------receipt------")

total=total_cost(cost)

print(f"total cost = ${total:.2f}") # prints total cost to 2dp

discounted_price=discount(total)

print(f"discounted cost = ${discounted_price:.2f}") #prints discounted price

print(f"rewards point = {int(rewards_point(discounted_price))}") #prints rewards

if discounted_price < 25:
  print("You need to sped over $25 for a voucher code.")
else:
  code=vouchers(discounted_price,name)
  print(f"your voucher code is {code}")

with open ("vouchercode.txt","w") as file: #file automatically closes after the with block
  file.write(code)
```
