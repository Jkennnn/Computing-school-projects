```python
word=input("Enter a word: ")
while True:
  if len(word)==6:
    if word.isalpha():
      break
  print("Word can only contain 6 letters and word can only be letters")
  word=input("Enter a word: ")

count=0
answer=["_", "_", "_","_","_","_"]
while "_" in answer:
  guess=input("Guess a letter: ")
  while True:
    if len(guess)==1 and guess.isalpha():
      break
    print("guess is invalid")
    guess=input("Guess a letter: ")
  for i in range(len(word)):
    if guess==word[i]:
      answer[i]=guess
  print(answer)
  count+=1

print("you took", count,  "guesses")
```
